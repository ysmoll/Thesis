```python
import duckdb
import pandas as pd
```

```python
def create_con():
    con = duckdb.connect(database=':memory:')
    return con

def execute_local(con, query):
    result = con.execute(query).fetchdf()
    return result

con = create_con()
q = '''
        CREATE TABLE patients AS SELECT * FROM read_csv_auto('Data/synthetic/csv/patients.csv')
    '''
execute_local(con, q)
patients = execute_local(con, '''select * from patients limit 10''')
patients
```

```python
def distinguishable(x, y):
    if not len(x) == len(y):
        return True
    if x.equals(y):
        return False
    return True

def validate_k_anonimity(df):

    for i in range(len(df)):
        for j in range(len(df)):
            if i == j:
                continue
            current = df.iloc[i]
            
            print(distinguishable(current, current))
            
            return

validate_k_anonimity(patients)
```
